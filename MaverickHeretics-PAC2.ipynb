```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob 

# Carreguem el conjunt de dades deTwitter US Airline Sentiment
df = pd.read_csv('Tweets.csv')

# Realitzem l'anàlisi de sentiments mitjançant TextBlob
df['Sentiment'] = df['text'].apply(lambda x: TextBlob(x).sentiment.polarity)

# Visualitzem la distribució de sentiments
plt.figure(figsize=(10, 6))
sns.histplot(df['Sentiment'], bins=30, kde=True, color='skyblue')
plt.title('Distribució de sentiment dels tweets')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.show()
```
